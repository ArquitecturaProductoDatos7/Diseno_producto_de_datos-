```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#res = requests.get('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=incidentes-viales-c57&facet=hora_creacion&facet=dia_semana&facet=codigo_cierre&facet=mesdecierre&facet=delegacion_inicio&facet=incidente_c4&facet=clas_con_f_alarma&facet=tipo_entrada&facet=mes&facet=folio&facet=fecha_creacion&facet=fecha_cierre&facet=hora_cierre&refine.fecha_creacion=04%2F09%2F2015').json()['records']
res = requests.get('https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=incidentes-viales-c5&refine.fecha_creacion=04%2F09%2F2015').json()['records']
```

```python
tabla = []

for f in res:
    data = f['fields']
    folio = data['folio']
    fecha_creacion = data['fecha_creacion']
    hora_creacion = data['hora_creacion']
    latitud = data['geopoint'][0]
    longitud = data['geopoint'][1]
    delegacion_inicio = data['delegacion_inicio']
    tipo_entrada = data['tipo_entrada']
    incidente = data['incidente_c4']
    clasificacion = data['clas_con_f_alarma']
    codigo_cierre = data['codigo_cierre']
    fecha_cierre = data['fecha_cierre']
    hora_cierre = data['hora_cierre']
    fila = (folio,fecha_creacion,hora_creacion,latitud,longitud,delegacion_inicio,tipo_entrada,incidente,clasificacion,
            codigo_cierre,fecha_cierre,hora_cierre)
    tabla.append(fila)
    
    
```

```python
df = pd.DataFrame(tabla, columns = ['folio','fecha_creacion','hora_creacion','latitud','longitud','delegacion_inicio',
                                  'tipo_entrada','incidente','clasificacion','codigo_cierre','fecha_cierre','hora_cierre'])
```
